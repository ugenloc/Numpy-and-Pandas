## Introduction to pandas

pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language. 



```python
import pandas as pd

# Create a series in pandas
make = pd.Series(["Toyota","Honda", "Toyota", "BMW", "Nissan", "Toyota", "BMW", "Nissan", "Toyota", "Honda", "Honda", "Toyata", "Nissan"])
make
```

```python
# Series are one dimensional data
colors = pd.Series(["White", "Red", "Blue", "Black", "White", "Green", "Blue", "Blue", "White", "White"])
Odometer = pd.Series([150043, 87899, 32549, 11179, 213095, 99213, 45698, 54738, 60000, 31600])  
doors = pd.Series([4, 4, 3, 5, 4, 4, 4, 4, 4, 4])
price = pd.Series([4000, 5000, 7000, 22000, 3500, 4500, 7500, 7000, 6500, 97700])
```

```python
# DataFrame are 2-Dimentional and a series can be passed into a DataFrame

car_sales= pd.DataFrame({"make": make, "colors": colors, "Odometer (KM)": Odometer, "Doors": doors, "Price": price})
car_sales
```

# Manipulating Data

```python
#String method - this changes all the strings to lower case
car_sales["make"].str.lower()
```

```python
# Let's see if our lower case string saved
car_sales
```

```python
# This is the correct way to re-aasign and save string case
car_sales["make"] = car_sales["make"].str.lower()
car_sales
```

```python
# Use fillna() function to fill missing values -- Fill a particular column with it's  own mean
car_sales["Odometer (KM)"].fillna(car_sales["Odometer (KM)"].mean())
```

```python
car_sales["Odometer (KM)"].mean()
```

```python
# Set inplace parameter inside fillna function to true so as to replace the missing values
car_sales["Odometer (KM)"].fillna(car_sales["Odometer (KM)"].mean(), inplace=True)
```

```python
# Re-assign car sales to a new variable and  Use dropna function to get rid if missing data 
carSalesDropedNa = car_sales.dropna()
carSalesDropedNa
```

```python
carSalesDropedNa.to_csv("carSalesDropedNa.csv")
```

```python
car_sales
```

```python
# Create a Column from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 7.5, 9.2, 5.0, 9.6, 8.7, 3.0, 2.0, 1.0]
car_sales["Fuel per 100km"] = fuel_economy
car_sales
```

```python
# Create a new column from existing column
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100km"]
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales.dtypes
```

```python
# Drop a colum -- the drop function requires axis to be specified
car_sales = car_sales.drop("Total fuel used", axis=1)
car_sales
```

```python
# Anatomy of DataFrame
# Column axis = 1, Row axis = 0
```

```python
# Shuffling rows in a  dataframe
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Only select 20% of the data
car_sales_shuffled.sample(frac=0.2)
```

```python
# re-set index
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

```python
# Convert Odometer KM to miles by dividing by 1.6
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
